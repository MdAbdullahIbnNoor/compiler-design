```python
def analyze_c_code_final(c_code):
    identifiers = set()
    operators = set()
    constants = set()
    keywords = set()
    special_symbols = set()

    i = 0
    while i < len(c_code):
        char = c_code[i]

        # Skip whitespaces
        if char.isspace():
            i += 1
            continue

        # Check for identifiers or keywords
        if char.isalpha() or char == '_':
            identifier = char
            i += 1
            while i < len(c_code) and (c_code[i].isalnum() or c_code[i] == '_'):
                identifier += c_code[i]
                i += 1

            if identifier in ["auto","double","int","struct","break","else","long","switch"
            ,"case","enum","register","typedef","char","extern","return","union" ,"include",
            "continue","for","signed","void","do","if","static","while","endif","then"
            ,"default","goto","sizeof","volatile","const","float","short","unsigned", "printf","scanf", "case", "char", "const", "continue", "default", "do", "double", "else", "enum", "extern", "float", "for", "goto", "if", "int", "long", "register", "return", "short", "signed", "sizeof", "static", "struct", "switch", "typedef", "union", "unsigned", "void", "volatile", "while"]:
                keywords.add(identifier)
            else:
                identifiers.add(identifier)
            continue

        # Check for operators
        if char in ["+", "-", "*", "/", "%", "=", "==", "!=", "<", ">", "<=", ">=", "&&", "||", "!", "&", "|", "^", "~", "<<", ">>", "+=", "-=", "*=", "/=", "%=", "<<=", ">>=", "&=", "^=", "|="]: 
            operators.add(char)
            i += 1
            continue
          

        # Check for constants
        if char.isdigit():
            constant = char
            i += 1
            while i < len(c_code) and (c_code[i].isdigit() or c_code[i] == '.'):
                constant += c_code[i]
                i += 1
            constants.add(constant)
            continue

        # Check for special symbols
        if not char.isalnum() and not char.isspace():
            special_symbols.add(char)
            i += 1
            continue

        i += 1

    # Display the results
    print("Identifiers:", list(identifiers))
    print("Keywords:", list(keywords))
    print("Operators:", list(operators))
    print("Constants:", list(constants))
    print("Special Symbols:", list(special_symbols))

# Example usage
c_code_input =input()


analyze_c_code_final(c_code_input)
```
